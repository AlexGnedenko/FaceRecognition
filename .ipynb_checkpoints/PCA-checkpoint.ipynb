### Подключаем библиотеки:

```python
import numpy as np
import math
```

### Генерация пробной матрицы данных:

```python
matrix = [[1,2,3,4,5,6,7,8,9],
          [9,8,7,6,5,4,3,2,1],
          [10,20,30,40,50,60,70,80,90]]
```

### Центрирование базы данных:

```python
matrix = matrix - np.mean(matrix)
m = matrix.mean()
print(matrix)
```

### Ковариационная матрица:

```python
covMatrix = np.dot(matrix.transpose(),matrix)
print(covMatrix)
```

### Собственные значения и вектора:

```python
eigenvalues, eigenvectors  = np.linalg.eig(covMatrix)
print(eigenvalues)
print(eigenvectors)
```
